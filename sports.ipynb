```
## отчистим данные
import csv

lst = []

with open('sport.stat', 'r') as file:
    sport = csv.reader(file, delimiter='\t')
    for line in sport:
      if line[0] == 'Номер':
          continue
      else:
          #print(line[3])
          if not line[3].strip() == '':
              #print(line[3].strip().split())
              for i in line[3].strip().split(','):
                  lst.append(i.strip())
import collections

count = collections.Counter(lst)
dict(count)
for s in sorted(dict(count), key=count.get, reverse = True):
    print(s, count[s])
```
